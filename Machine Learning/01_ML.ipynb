```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("D:\\AI & ML\\Pandas\\01_ML.csv")
df.head()
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['year ']],df['per capita income(US$)'])
```

```python
reg.predict(np.array([[2020]])) # predict takes a 2D array not a normal array..
```

```python
plt.plot(df['year '],df['per capita income(US$)'],linestyle = 'none',marker = '+')
plt.xlabel("year")
plt.ylabel("per capita income(US$)")
plt.plot(df['year '],reg.predict(df[['year ']]),color="blue")
```
