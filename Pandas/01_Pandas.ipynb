```python
import pandas as pd
```

```python
# DataFrame Basics

# Read a CSV file -- 
df = pd.read_csv("D:\\AI & ML\\Pandas\\Book2.csv")
df
```

```python
#Want to get only few tables from top rather than the whole table use .head() function.. 
df.head()
#You can pass value in .head() [ like .head(2) , to get only 2 rows from top ] .. 
df.head(2)
```

```python
# Similarly you can use .tail() function which works exactly opposite to .head()
#Want to get only few tables from bottom rather than the whole table use .tail() function.. 
df.tail()
#You can pass value in .tail() [ like .tail(2) , to get only 2 rows from bottom ] .. 
df.tail(2)
```

```python
# To get the no. of rows and columns from a database, you can use .shape [it's not a function it's a property, returns a "tuple"]
df.shape 
rows,columns = df.shape
print(rows, columns)
```

```python
# You can also use indexing and slicing techiniques to print specific rows [ like data from row 2 to 5 ] 
df[2:5] # Here 2nd rows of df included but 5th row not included [ works for 2nd,3rd and 4th rows ] 
```

```python
# To print a specific column details, you can get it by just it's name
df.Temperature # or use , df['Temperature']
```

```python
df[['Temperature', 'Humidity' ]]
```

```python
# .max() function is used to get the maximum integer value from a specific column.
df['Humidity'].max()
```

```python
# Similarly , 
# .min() function is used to get the minimum integer value from a specific column.
df['Humidity'].min()
```

```python
# To get the Average of a particular column,
# .mean() function is used to get the average integer value of a specific column.
df['Humidity'].mean()
```

```python
# You can also define conditions to get a specific result [ not exactly but like SQL Queries ]..
df.loc[df.Humidity >= 21,['Temperature','Humidity','Events']]
# One more example
df.loc[df.Humidity == df.Humidity.max(),['Temperature','Humidity','Events']]
```
