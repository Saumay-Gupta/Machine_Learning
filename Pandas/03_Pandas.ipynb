```python
# Suppose there is no header in your DataFrame and you want to create a header..
# 1st use "header=None" to place a temperary header initialised with indexes(0,1,2..)
# 2nd use "names = []" to give the name of columns in place of those indexes..
df = pd.read_csv("D:\\AI & ML\\Pandas\\Book2.csv",header=None,names=["Day","Temperature","Windspeed","Event"])
df
```

```python
# If you want to read only limited data [i.e some few rows], use "nrows" to define no. of rows to read..
df = pd.read_csv("D:\\AI & ML\\Pandas\\Book2.csv",nrows = 2)
df
```

```python
# Now suppose you have a some messy data [i.e some entries are empty (like not availabe, n.a.)], your task to clean that Data.
# Use "na_values" to declare "NaN" to those column you want.. 
df = pd.read_csv("D:\\AI & ML\\Pandas\\Book2.csv",na_values={
    'eps': ["not available", "n.a."],
    'revenue': ["not available", "n.a.",-1],
    'price': ["not available", "n.a.",-1],
    'people': ["not available", "n.a."]
})
df
```

```python
# To change a specific value of a cell, we can use "converters"..

def convert_people_cell(name):
    if name == "n.a.":
            return "sam walton"
    return name
    
df = pd.read_csv("D:\\AI & ML\\Pandas\\Book2.csv",converters={
    'people': convert_people_cell
})

# OR use below method as well to change value of a specific column..

df['people'] = df['people'].replace("n.a.","Sam walton")

df
```
