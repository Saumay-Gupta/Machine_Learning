```python
# Use "interpolate()" method is used to fill values(NaN) by intermediate values..
# By default interpolate() works on linear method..[i.e fill (NaN) values by calculating the middle value]
# new_df = df.interpolate()
# new_df

# interpolate() also has some other methods like "time" [ that guess/calculate values based on the Data&Time].. 
new_df = df.interpolate(method="time")
new_df
```

```python
# You can drop the rows which contains NaN values, by using "dropna()" method..
# new_df = df.dropna() # This will drop all rows which have atleast 1 NaN cell/column.. 
# new_df

# use "thresh" argument in "dropna()" to specify condition [like thresh=1 means include rows, which contains atleast 1 column with data(not with NaN)]..
new_df = df.dropna(thresh=1)
new_df
```
