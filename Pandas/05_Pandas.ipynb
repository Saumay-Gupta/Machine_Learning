```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:\\AI & ML\\Pandas\\Tut_05.csv")
df
```

```python
# To replace all "-99999" with "NaN" we use ".replace()" function..
# .replace() takes 2 argument, 1st is  the value which you want to replace 2nd is the value with which you want to replace..
# we can pass a list,tuple,dictionary as the 1st argument..
new_df = df.replace({
    'temperature':-99999,
    'windspeed':-99999,
    'event': '0'
    },np.nan) # "np.nan" is a pre-built keyword in Numpy, which means NaN value..
new_df
```

```python
# But let's assume we want to declare different values to different columns..
# For the value "-99999" replace with "NaN" and 0 with "Sunny"
# For this we can also use ".replace()" ..
new_df = df.replace({
    '-99999':np.nan,
    '0':"Sunny"
})
new_df
```

```python
# Now we want to extact the units [like temperature units(C,F) and windspeed units(mph)]..
# We use Regular Expression with .replace() ..
import re
new_df2 = new_df.replace({
    'temperature': '[A-Za-z]',
    'windspeed': '[A-Za-z]'
},'',regex=True)
new_df2
```
