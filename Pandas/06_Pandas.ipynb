```python
import pandas as pd
```

```python
df = pd.read_csv("D:\\AI & ML\\Pandas\\Tut_06.csv")
df
```

```python
g = df.groupby('city') # ".groupby()" methods is used to make group the data of similar entities based of the column, you provide..
g # ".groupby()" return an object
```

```python
for city,city_df in g:  # Object returned by ".groupby()" contains column name and the DataFrame, that you can use like this using iterator..
    print(city)
    print(city_df)
```

```python
g.get_group('mumbai') # To access the individual DataFrame or a single group, use ".get_group()" method, it also returns a DataFrame 
```

```python
g.max() #This will perform functions on each individual DataFrame..
```

```python
g.describe() #This will perform functions on each individual DataFrame..
```
